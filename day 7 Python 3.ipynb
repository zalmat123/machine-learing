```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/mtcars.csv")
```

```python
plt.scatter(df['hp'], df['cyl'])
plt.xlabel("HP")
plt.ylabel("CYL")
plt.title("HP vs Cyl")
plt.show()
```

```python
plt.scatter(df['mpg'], df['hp'])
plt.xlabel("MPG")
plt.ylabel("HP")
plt.title("MPG vs HP")
plt.show()
```

```python
name = ["A","B","C","D","X","Y","Z"]
salary = np.random.randint(30000, 80000,7)
salary
```

```python
color_name = ["r", "g", "b", "k", "#acd448", "#46216e", "#ffaa00"]
```

```python
color_name
```

```python
plt.bar(name, salary, color = color_name)
plt.xlabel("Name")
plt.ylabel("Salary")
plt.title("Name vs Salary")
plt.show()
```
