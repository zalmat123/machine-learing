```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# step1: take the data
df = pd.read_csv("/content/LSM1.csv")
```

```python
# step2: filter the data
# step3: divide into the input and output, x-->input, y-->output
# retrive the x and y value from the dataframe
x = df.iloc[:,0].values.reshape(5,1)
y = df.iloc[:,1].values
```

```python
# create an object  since we cant use the a class directly
Lin = LinearRegression()
```

```python
# training the model
Lin.fit(x,y)
```

```python
Lin.predict([[2.5]])
```
