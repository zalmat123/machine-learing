```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("/content/LSM1.csv")
```

```python
x = df.iloc[:,0].values.reshape(5,1)
```

```python
y = df.iloc[:,1].values
```

```python
from sklearn.linear_model import LinearRegression
Lin = LinearRegression()
Lin.fit(x,y)
Lin.predict([[2.5]])
```

```python
Lin.predict(x)
```

```python
Lin.coef_
```

```python
Lin.intercept_
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x, y,color="r", marker="*")
plt.plot(x, Lin.predict(x), color ="g", marker ="*", markersize = 10)
plt.show()
```
