Finding the missing value

```python
# missing value 
# create the data frame
# np.nan ==> not a number
z1=[1,np.nan,np.nan,4,5,np.nan,7,np.nan]
z2=[1,1,1,1,2,2,2,2]
#the above two z are a list
# now create the dataframe the like the following
df1 = pd.DataFrame({"Cricket":z1,
                   "Tennis":z2})
```

```python
df1
```

```python
df1.fillna(value=0)
```

```python
df1
```

```python
df1.fillna(2)
```

```python
df1.fillna(4.25)
```

```python
df1['Cricket'].mean()
```

```python
df1
```

```python
df1.fillna(df1['Cricket'].mean())
```

```python
df1[:4].mean()
```

```python
df1['Cricket'].fillna(df1.groupby(['Tennis'])['Cricket'].transform(np.mean))
```
